```python
<!DOCTYPE html>
<html>
<head>
 <title>Front-End Page</title>
 <style>
 body {
 font-family: Arial, sans-serif;
 margin: 0;
 padding: 0;
 background-color: #f0f0f0;
 }

 h1 {
 font-size: 2em;
 padding: 20px;
 background-color: #333;
 color: #fff;
 }

 .input-group {
 display: flex;
 flex-direction: column;
 margin: 20px;
 }

 label {
 font-size: 1.2em;
 margin-bottom: 10px;
 }

 input[type="file"] {
 border: 1px solid #ccc;
 padding: 10px;
 font-size: 1em;
 }

 button {
 background-color: #4CAF50;
 color: white;
 padding: 10px;
 border: none;
 font-size: 1.2em;
 cursor: pointer;
 margin-top: 10px;
 }

 .error-message {
 color: red;
 font-size: 1em;
 margin-top: 10px;
 }

 .download-template-button {
 margin-top: 20px;
 }

 .notification-message {
 color: green;
 font-size: 1em;
 margin-top: 10px;
 }
 </style>
</head>
<body>
 <h1>Front-End Page</h1>
 <div class="input-group">
 <label for="file-input">Select a file:</label>
 <input type="file" id="file-input">
 <button id="load-button">Load</button>
 <div id="error-message" class="error-message"></div>
 </div>
 <div class="download-template-button">
 <button id="download-template-button">Download Template</button>
 </div>
 <div id="notification-message" class="notification-message"></div>
 <script>
 document.getElementById("load-button").addEventListener("click", function() {
 const fileInput = document.getElementById("file-input");
 if (fileInput.files.length === 0) {
 document.getElementById("error-message").innerText = "Please select a file.";
 } else {
 const formData = new FormData();
 formData.append('file', fileInput.files[0]);

 fetch('/upload', {
 method: 'POST',
 body: formData
 }).then(response => response.text())
 .then(data => {
 document.getElementById("notification-message").innerText = data;
 })
 .catch(error => {
 console.error('Error:', error);
 });
 }
 });

 document.getElementById("download-template-button").addEventListener("click", function() {
 window.location.href = '/download_template';
 });
 </script>
</body>
</html>
```
