```python
import pdfkit
import requests

def convert_html_to_pdf(input_path, output_path):
    """Converts HTML content to PDF, preserving design as much as possible.

    Args:
        input_path (str): Path to the HTML file or web URL.
        output_path (str): Path to save the generated PDF.
    """

    try:
        if input_path.startswith("http"):
            # Fetch HTML content from URL
            response = requests.get(input_path)
            content = response.content.decode('utf-8')  # Decode bytes to string
            
            # Save HTML content locally (optional)
            with open("temp.html", "w", encoding="utf-8") as f:
                f.write(content)

            # Convert local HTML file to PDF
            pdfkit.from_file("temp.html", output_path, options={'no-images': None})
        else:
            # Read HTML content from local file
            with open(input_path, "r", encoding="utf-8") as f:
                content = f.read()

            # Use pdfkit to convert HTML to PDF, handling potential errors
            pdfkit.from_string(content, output_path, options={'no-images': None})

        print(f"HTML content successfully converted to PDF and saved at: {output_path}")

    except Exception as e:
        print(f"Error converting HTML to PDF: {e}")

# Example usage:
input_path ="https://www.ti.com/product/ADS8664?keyMatch=ADS8664IDBT&tisearch=Search-EN-Everything"  # Replace with your desired HTML path or URL
output_path = "output.pdf"  # Replace with your desired output PDF path
convert_html_to_pdf(input_path, output_path)
```

```python
import fitz

def highlight_text(pdf_path, search_term, highlight_color="#FFFF00"):
    """Highlights the given search term in a PDF."""
    doc = fitz.open(pdf_path)
    for page in doc:
        text = page.get_text("text")
        if search_term in text:
            for block in page.get_text("blocks"):
                if search_term in block[4]:
                    rect = fitz.Rect(block[0], block[1], block[2], block[3])
                    annot = page.add_highlight_annot(rect, color=highlight_color)
    doc.save("highlighted_pdf.pdf")

# Get user input
user_input = input("Enter the word to highlight in the PDF: ")
pdf_path = r'C:\python files\Screenshort higlight\output1.pdf'

# Highlight the word
highlight_text(pdf_path, user_input)

print("Word highlighted! Output saved as 'highlighted_pdf.pdf'")
```

```python
import fitz
def highlight_and_display(input_pdf, output_pdf, search_text):
    doc = fitz.open(input_pdf)
    page = doc[0]  # Assuming you want to highlight text on the first page
    text_instances = page.search_for(search_text)

    for inst in text_instances:
        highlight = page.add_highlight_annot(inst)
        highlight.update()

    doc.save(output_pdf)
    doc.close()

# Example usage:
user_input_word = input("Enter the word to highlight: ")
highlight_and_display("output1.pdf", "hili.pdf", user_input_word)


# ----------------------------------------------------------------------------working-----------------------------------------------------------------------------------------------------
```

```python
import bs4

def highlight_word(html_file, word):
    """Highlights a given word in an HTML file and saves the modified file."""

    with open(html_file, 'r',encoding='Utf-8') as f:
        soup = bs4.BeautifulSoup(f, 'html.parser')

    for text_node in soup.find_all(text=True):
        if word in text_node:
            new_text = text_node.replace(word, f'<mark>{word}</mark>')
            text_node.replace_with(new_text)
    with open(html_file, 'w', encoding='utf-8') as f:  # Specify encoding for writing
        f.write(str(soup))

if __name__ == '__main__':
    html_file = r'C:\python files\Screenshort higlight\temp.html'  # Replace with the actual path
    word_to_highlight = input("Enter the word to highlight: ")
    highlight_word(html_file, word_to_highlight)
    print("Word highlighted successfully!")
```

```python
import pdfkit
import requests
from bs4 import BeautifulSoup
from weasyprint import HTML

def fetch_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for non-200 status codes

        soup = BeautifulSoup(response.text, 'html.parser')

        # Remove unwanted elements or apply CSS styling (optional)
        if True:  # Replace with your filtering/styling logic
            for script in soup.find_all('script'):
                script.decompose()  # Remove scripts
            for style in soup.find_all('style'):
                style.decompose()  # Remove inline styles

        return soup.prettify()  # Indented HTML for better readability

    except requests.exceptions.RequestException as e:
        print(f"Error fetching content: {e}")
        return None

def convert_to_pdf(html_content, output_file_name):
    try:
        # Use WeasyPrint for better quality and handling of complex HTML
        pdfkit.from_string(html_content, output_file_name, options={'quiet': True})

    except Exception as e:
        print(f"Error generating PDF: {e}")

def main():
    url = input("Enter the URL of the HTML page: ")
    html_content = fetch_content(url)

    if html_content:
        output_file_name = url.split('/')[-1] + ".pdf"
        convert_to_pdf(html_content, output_file_name)
        print(f"PDF generated successfully: {output_file_name}")
    else:
        print("Error fetching content.")

if __name__ == "__main__":
    main()


# ----------------------------------------------------------------------------working-----------------------------------------------------------------------------------------------------
```
